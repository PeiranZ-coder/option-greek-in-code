```python
"""
option_greeks.py
Black-Scholes-Merton option Greeks (with continuous dividend yield q).

Direct Python port of option_greeks.cpp / option_greeks.h, but fully
vectorized: any of S, K, T, sigma, r, q may be a NumPy array, so you can
sweep a parameter and get a whole curve back in one call — ideal for plotting.
"""
from dataclasses import dataclass
import numpy as np
from scipy.stats import norm

__all__ = ["OptionParams", "d1_d2", "bs_price",
           "delta", "gamma", "vega", "theta", "rho"]


@dataclass
class OptionParams:
    """Mirror of the C++ OptionParams struct."""
    S: float          # spot price of the underlying
    K: float          # strike
    T: float          # time to maturity (years)
    sigma: float      # volatility
    r: float          # risk-free rate (continuous)
    q: float = 0.0    # dividend yield (0 if none)


# ---------- d1 and d2 ----------
def d1_d2(p: OptionParams):
    sig_sqrt_T = p.sigma * np.sqrt(p.T)
    d1 = (np.log(p.S / p.K) + (p.r - p.q + 0.5 * p.sigma ** 2) * p.T) / sig_sqrt_T
    d2 = d1 - sig_sqrt_T
    return d1, d2


# ---------- Black-Scholes price (bonus: not in the C++, handy for plots) ----------
def bs_price(p: OptionParams, is_call: bool = True):
    d1, d2 = d1_d2(p)
    disc_r = np.exp(-p.r * p.T)
    disc_q = np.exp(-p.q * p.T)
    if is_call:
        return p.S * disc_q * norm.cdf(d1) - p.K * disc_r * norm.cdf(d2)
    return p.K * disc_r * norm.cdf(-d2) - p.S * disc_q * norm.cdf(-d1)


# ---------- 1. Delta ----------
def delta(p: OptionParams, is_call: bool = True):
    d1, _ = d1_d2(p)
    df = np.exp(-p.q * p.T)
    return df * norm.cdf(d1) if is_call else df * (norm.cdf(d1) - 1.0)


# ---------- 2. Gamma (same for call & put) ----------
def gamma(p: OptionParams):
    d1, _ = d1_d2(p)
    return np.exp(-p.q * p.T) * norm.pdf(d1) / (p.S * p.sigma * np.sqrt(p.T))


# ---------- 3. Vega (same for call & put) ----------
def vega(p: OptionParams):
    d1, _ = d1_d2(p)
    return p.S * np.exp(-p.q * p.T) * norm.pdf(d1) * np.sqrt(p.T)
    # per 1% vol move:  divide by 100


# ---------- 4. Theta (annualized) ----------
def theta(p: OptionParams, is_call: bool = True):
    d1, d2 = d1_d2(p)
    term1 = -(p.S * p.sigma * np.exp(-p.q * p.T) * norm.pdf(d1)) / (2.0 * np.sqrt(p.T))
    if is_call:
        term2 = -p.r * p.K * np.exp(-p.r * p.T) * norm.cdf(d2)
        term3 =  p.q * p.S * np.exp(-p.q * p.T) * norm.cdf(d1)
    else:
        term2 =  p.r * p.K * np.exp(-p.r * p.T) * norm.cdf(-d2)
        term3 = -p.q * p.S * np.exp(-p.q * p.T) * norm.cdf(-d1)
    return term1 + term2 + term3
    # per calendar day:  divide by 365


# ---------- 5. Rho ----------
def rho(p: OptionParams, is_call: bool = True):
    _, d2 = d1_d2(p)
    if is_call:
        return p.K * p.T * np.exp(-p.r * p.T) * norm.cdf(d2)
    return -p.K * p.T * np.exp(-p.r * p.T) * norm.cdf(-d2)
    # per 1% rate move:  divide by 100


if __name__ == "__main__":
    # quick self-check against a single point (matches the C++ output)
    p = OptionParams(S=100, K=100, T=1.0, sigma=0.20, r=0.05, q=0.0)
    print(f"Call Delta : {delta(p, True):.6f}")
    print(f"Put  Delta : {delta(p, False):.6f}")
    print(f"Gamma      : {gamma(p):.6f}")
    print(f"Vega       : {vega(p):.6f}   (per 1.00 vol; /100 for 1%)")
    print(f"Call Theta : {theta(p, True):.6f}   (annual; /365 for daily)")
    print(f"Put  Theta : {theta(p, False):.6f}")
    print(f"Call Rho   : {rho(p, True):.6f}   (per 1.00 rate; /100 for 1%)")
    print(f"Put  Rho   : {rho(p, False):.6f}")
    print(f"Call price : {bs_price(p, True):.6f}")
    print(f"Put  price : {bs_price(p, False):.6f}")
```

```python
"""
plot_greeks.py
Plot the option Greeks as a function of spot price.

Because option_greeks.py is vectorized, plotting is just:
  1. build an array of spot prices,
  2. put it in OptionParams,
  3. call the Greek functions,
  4. plot.

Change `base` below (K, T, sigma, r, q) and re-run to explore.
"""
import numpy as np
import matplotlib.pyplot as plt
from option_greeks import OptionParams, delta, gamma, vega, theta, rho, bs_price

# ---- market setup -----------------------------------------------------
K, T, sigma, r, q = 100.0, 0.5, 0.20, 0.03, 0.0
S = np.linspace(60, 140, 400)
p = OptionParams(S=S, K=K, T=T, sigma=sigma, r=r, q=q)

CALL, PUT, ONE = "#12897E", "#E76F51", "#3B6FB6"


def _fmt(ax, title, ylabel, atm=True):
    ax.set_title(title, fontsize=12, fontweight="bold")
    ax.set_xlabel("Spot price  S")
    ax.set_ylabel(ylabel)
    ax.grid(True, color="#E1E6EF", lw=0.8)
    for s in ("top", "right"):
        ax.spines[s].set_visible(False)
    if atm:
        ax.axvline(K, color="#888", ls="--", lw=1)
    ax.axhline(0, color="#333", lw=0.7)


fig, axes = plt.subplots(2, 3, figsize=(15, 8.5))
fig.suptitle(f"Option Greeks  (K={K}, T={T}y, σ={sigma}, r={r}, q={q})",
             fontsize=15, fontweight="bold")

# Delta
ax = axes[0, 0]
ax.plot(S, delta(p, True),  color=CALL, lw=2.2, label="Call")
ax.plot(S, delta(p, False), color=PUT,  lw=2.2, label="Put")
_fmt(ax, "Delta  Δ = ∂V/∂S", "Delta"); ax.legend(frameon=False)

# Gamma
ax = axes[0, 1]
ax.plot(S, gamma(p), color=ONE, lw=2.2)
_fmt(ax, "Gamma  Γ = ∂²V/∂S²", "Gamma")

# Vega
ax = axes[0, 2]
ax.plot(S, vega(p) / 100, color=ONE, lw=2.2)
_fmt(ax, "Vega  ν = ∂V/∂σ", "Vega  (per 1% vol)")

# Theta
ax = axes[1, 0]
ax.plot(S, theta(p, True)  / 365, color=CALL, lw=2.2, label="Call")
ax.plot(S, theta(p, False) / 365, color=PUT,  lw=2.2, label="Put")
_fmt(ax, "Theta  Θ = ∂V/∂t", "Theta  (per day)"); ax.legend(frameon=False)

# Rho
ax = axes[1, 1]
ax.plot(S, rho(p, True)  / 100, color=CALL, lw=2.2, label="Call")
ax.plot(S, rho(p, False) / 100, color=PUT,  lw=2.2, label="Put")
_fmt(ax, "Rho  ρ = ∂V/∂r", "Rho  (per 1% rate)"); ax.legend(frameon=False)

# Price (bonus panel)
ax = axes[1, 2]
ax.plot(S, bs_price(p, True),  color=CALL, lw=2.2, label="Call")
ax.plot(S, bs_price(p, False), color=PUT,  lw=2.2, label="Put")
_fmt(ax, "Option price  V", "Price"); ax.legend(frameon=False)

fig.tight_layout(rect=[0, 0, 1, 0.96])
fig.savefig("greeks_plot.png", dpi=150, bbox_inches="tight")
print("saved greeks_plot.png")
```

```python
"""
greeks_plot_standalone.py
Black-Scholes-Merton Greeks (with dividend yield q) + plots, all in ONE file.
No local imports to resolve — just run:  python greeks_plot_standalone.py

Requires: numpy, scipy, matplotlib
    pip install numpy scipy matplotlib
"""
from dataclasses import dataclass
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# ======================================================================
#  Greeks  (vectorized: S, K, T, sigma, r, q may be scalars or arrays)
# ======================================================================
@dataclass
class OptionParams:
    S: float          # spot
    K: float          # strike
    T: float          # time to maturity (years)
    sigma: float      # volatility
    r: float          # risk-free rate
    q: float = 0.0    # dividend yield


def d1_d2(p):
    sig_sqrt_T = p.sigma * np.sqrt(p.T)
    d1 = (np.log(p.S / p.K) + (p.r - p.q + 0.5 * p.sigma ** 2) * p.T) / sig_sqrt_T
    d2 = d1 - sig_sqrt_T
    return d1, d2


def bs_price(p, is_call=True):
    d1, d2 = d1_d2(p)
    dr, dq = np.exp(-p.r * p.T), np.exp(-p.q * p.T)
    if is_call:
        return p.S * dq * norm.cdf(d1) - p.K * dr * norm.cdf(d2)
    return p.K * dr * norm.cdf(-d2) - p.S * dq * norm.cdf(-d1)


def delta(p, is_call=True):
    d1, _ = d1_d2(p)
    df = np.exp(-p.q * p.T)
    return df * norm.cdf(d1) if is_call else df * (norm.cdf(d1) - 1.0)


def gamma(p):
    d1, _ = d1_d2(p)
    return np.exp(-p.q * p.T) * norm.pdf(d1) / (p.S * p.sigma * np.sqrt(p.T))


def vega(p):                       # per 1.00 vol; /100 for 1%
    d1, _ = d1_d2(p)
    return p.S * np.exp(-p.q * p.T) * norm.pdf(d1) * np.sqrt(p.T)


def theta(p, is_call=True):        # annualized; /365 for per day
    d1, d2 = d1_d2(p)
    term1 = -(p.S * p.sigma * np.exp(-p.q * p.T) * norm.pdf(d1)) / (2.0 * np.sqrt(p.T))
    if is_call:
        term2 = -p.r * p.K * np.exp(-p.r * p.T) * norm.cdf(d2)
        term3 =  p.q * p.S * np.exp(-p.q * p.T) * norm.cdf(d1)
    else:
        term2 =  p.r * p.K * np.exp(-p.r * p.T) * norm.cdf(-d2)
        term3 = -p.q * p.S * np.exp(-p.q * p.T) * norm.cdf(-d1)
    return term1 + term2 + term3


def rho(p, is_call=True):          # per 1.00 rate; /100 for 1%
    _, d2 = d1_d2(p)
    if is_call:
        return p.K * p.T * np.exp(-p.r * p.T) * norm.cdf(d2)
    return -p.K * p.T * np.exp(-p.r * p.T) * norm.cdf(-d2)


# ======================================================================
#  Plot: every Greek vs spot price
# ======================================================================
def main():
    K, T, sigma, r, q = 100.0, 0.5, 0.20, 0.03, 0.0
    S = np.linspace(60, 140, 400)
    p = OptionParams(S=S, K=K, T=T, sigma=sigma, r=r, q=q)

    CALL, PUT, ONE = "#12897E", "#E76F51", "#3B6FB6"

    def fmt(ax, title, ylabel):
        ax.set_title(title, fontsize=12, fontweight="bold")
        ax.set_xlabel("Spot price  S"); ax.set_ylabel(ylabel)
        ax.grid(True, color="#E1E6EF", lw=0.8)
        for s in ("top", "right"):
            ax.spines[s].set_visible(False)
        ax.axvline(K, color="#888", ls="--", lw=1)
        ax.axhline(0, color="#333", lw=0.7)

    fig, ax = plt.subplots(2, 3, figsize=(15, 8.5))
    fig.suptitle(f"Option Greeks  (K={K}, T={T}y, sigma={sigma}, r={r}, q={q})",
                 fontsize=15, fontweight="bold")

    ax[0, 0].plot(S, delta(p, True),  color=CALL, lw=2.2, label="Call")
    ax[0, 0].plot(S, delta(p, False), color=PUT,  lw=2.2, label="Put")
    fmt(ax[0, 0], "Delta = dV/dS", "Delta"); ax[0, 0].legend(frameon=False)

    ax[0, 1].plot(S, gamma(p), color=ONE, lw=2.2)
    fmt(ax[0, 1], "Gamma = d2V/dS2", "Gamma")

    ax[0, 2].plot(S, vega(p) / 100, color=ONE, lw=2.2)
    fmt(ax[0, 2], "Vega = dV/dsigma", "Vega (per 1% vol)")

    ax[1, 0].plot(S, theta(p, True)  / 365, color=CALL, lw=2.2, label="Call")
    ax[1, 0].plot(S, theta(p, False) / 365, color=PUT,  lw=2.2, label="Put")
    fmt(ax[1, 0], "Theta = dV/dt", "Theta (per day)"); ax[1, 0].legend(frameon=False)

    ax[1, 1].plot(S, rho(p, True)  / 100, color=CALL, lw=2.2, label="Call")
    ax[1, 1].plot(S, rho(p, False) / 100, color=PUT,  lw=2.2, label="Put")
    fmt(ax[1, 1], "Rho = dV/dr", "Rho (per 1% rate)"); ax[1, 1].legend(frameon=False)

    ax[1, 2].plot(S, bs_price(p, True),  color=CALL, lw=2.2, label="Call")
    ax[1, 2].plot(S, bs_price(p, False), color=PUT,  lw=2.2, label="Put")
    fmt(ax[1, 2], "Option price V", "Price"); ax[1, 2].legend(frameon=False)

    fig.tight_layout(rect=[0, 0, 1, 0.96])
    fig.savefig("greeks_plot.png", dpi=150, bbox_inches="tight")
    print("saved greeks_plot.png")
    plt.show()   # opens a window; comment out if running headless


if __name__ == "__main__":
    main()
```
